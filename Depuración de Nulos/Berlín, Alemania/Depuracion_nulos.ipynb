```python
# Importar librer√≠as
import pandas as pd
import numpy as np
```

```python
# Cargar el dataset
df = pd.read_csv('base_berlin_filtrada.csv')
print(df.info())
```

```python
# Porcentaje Na's
# Porcentaje de NaN por columna
na_percent = df.isna().mean() * 100

# Ordenar de mayor a menor
na_percent = na_percent.sort_values(ascending=False)

print(na_percent)
```

```python
# Observar df
df
```

```python
# Observar valores nulos por columna
valores_nulos = df.isnull().sum()
valores_nulos_mayor_0 = valores_nulos[valores_nulos > 0]
valores_nulos_mayor_0
```

```python
# Primera y segunda sustituci√≥n valores nulos para host_name y host_since
df_host = df[df['host_name'].isna()].head(20)

null_host_ids = df_host['host_id'].unique()
print(null_host_ids)

# Verificar que los host_id con valores nulos en host_name y host_since tengan datos en otras columnas
check_null = df[(df['host_id'].isin(null_host_ids)) & (df['host_name'].notna()) & (df['host_since'].notna())]
print(check_null)

# Sustituci√≥n por el str Unknown para howst_name y host_since debido a la falta de relaci√≥n con otros host_id
df_clean = df.copy()
df_clean['host_name'] = df_clean['host_name'].fillna('Unknown')
df_clean['host_since'] = df_clean['host_since'].fillna('Unknown')

# Corroborar el cambio
print(df_clean.isnull().sum())
```

```python
# Sustituci√≥n para host_location y neighborhood
''' Se utilizar√° neighbourhood_cleansed para sustituir los valores nulos en host_location y neighborhood.'''
df_clean['host_location'] = df.apply(lambda row: row['neighbourhood_cleansed'].lower() if pd.isna(row['host_location']) else row['host_location'], axis=1)

df_clean['neighbourhood'] = df.apply(lambda row:row['neighbourhood_cleansed'].lower() if pd.isna(row['neighbourhood']) else row['neighbourhood'], axis = 1)
# Corroborar la sustituci√≥n
df_clean.isnull().sum()
```

```python
# Sustituci√≥n para host_response_time y host_response_rate

# Formato de host_response_rate y host_acceptance_rate a num√©rico
for col in ['host_response_rate', 'host_acceptance_rate']:
    df_clean[col] = pd.to_numeric(
        df_clean[col].astype(str).str.replace('%', '', regex=True),
        errors='coerce'
    )

""" Primero se sustituir√°n los valores nulos que est√©n en la misma fila de host_response_time por host_acceptance_rate. Luego."""
df_clean['host_response_rate'] = df_clean['host_response_rate'].fillna(df_clean['host_acceptance_rate'])
df_clean['host_response_time'] = df_clean['host_response_time'].fillna('No response')

# Sustituci√≥n para host_response_rate y host_acceptance_rate por un identificador num√©rico
df_clean['host_response_rate'].isin([88.88]).sum()
df_clean['host_acceptance_rate'].isin([89.89]).sum()

df_clean['host_response_rate'] = df_clean['host_response_rate'].fillna(88.88)
df_clean['host_acceptance_rate'] = df_clean['host_acceptance_rate'].fillna(89.89)

# Corroborar la sustituci√≥n
df_clean.isnull().sum()
```

```python
# Sustituci√≥n para los 211 na's de host_is_superhost utilizando el m√©todo backward fill
df_clean['host_is_superhost'] = df_clean['host_is_superhost'].fillna(method = 'bfill')
df_clean.isnull().sum()
```

```python
# Sustituci√≥n para host_neighbourhood y sus 8280 na's
'Se utilizar√° host_neighbourhood_cleansed para sustituir los valores nulos en host_neighbourhood.'
df_clean['host_neighbourhood'] = df_clean['host_neighbourhood'].fillna(df_clean['neighbourhood_cleansed'])

# Formato a host_neighbourhood y neighbourhood_cleansed
df_clean['host_neighbourhood'] = df_clean['host_neighbourhood'].str.lower().str.strip()
df_clean['neighbourhood_cleansed'] = df_clean['neighbourhood_cleansed'].str.lower().str.strip()

# Corroborar la sustituci√≥n
df_clean.isnull().sum()
```

```python
# Sustituci√≥n para los 13 na's host_listings_count, host_total_listings_count, host_verifications, host_has_profile_pic, host_identity_verified

# Debido a que no tienen otras propiedas registradas
df_clean['host_listings_count'] = df_clean['host_listings_count'].fillna(0)
df_clean['host_total_listings_count'] = df_clean['host_total_listings_count'].fillna(0)

# No se tiene informaci√≥n para sustituir
df_clean['host_verifications'] = df_clean['host_verifications'].fillna(method = 'bfill')

# Se asume que no tiene foto de perfil ni identidad verificada por falta de verificaci√≥n
df_clean['host_has_profile_pic'] = df_clean['host_has_profile_pic'].fillna('f')
df_clean['host_identity_verified'] = df_clean['host_identity_verified'].fillna('f')

# Corroborar la sustituci√≥n
df_clean.isnull().sum()


```

```python
# Sustituci√≥n para los 4955 na's de bathrooms_text
'''Se observ√≥ una relaci√≥n entre bathrooms_text y bathrooms. Se utilizar√° el n√∫mero que de bathrooms que hay en bathrooms_text para sustituir los valores nulos en bathrooms.'''
df_clean['bathrooms'] = df_clean['bathrooms'].fillna(df_clean['bathrooms_text'].str.extract(r'(\d+(\.\d+)?)')[0].astype(float))

# Corroborar la sustituci√≥n, aun restan 31 na's para bathroom y 7 na's para bathrooms_text
df_clean['bathrooms'] = df_clean['bathrooms'].fillna(0)
df_clean['bathrooms_text'] = df_clean['bathrooms_text'].fillna(method = "ffill")

df_clean.isnull().sum()
```

```python
# Sustituci√≥n para bedrooms y beds, apoyado con IA, por la importancia y representatividad de las variables.
''' Se utilizar√° la mediana de bedrooms y beds en base a la columna accommodates, con la que puede tener correlaci√≥n.'''
# Correlaci√≥n
print(df_clean[['accommodates','bedrooms','beds']].corr(method='pearson'))

#bedrooms
df_clean['bedrooms'] = df_clean['bedrooms'].fillna(
    df_clean.groupby('accommodates')['bedrooms'].transform('median')
).fillna(df_clean['bedrooms'].median())

#beds
df_clean['beds'] = df_clean['beds'].fillna(
    df_clean.groupby('accommodates')['beds'].transform('median')
).fillna(df_clean['bedrooms']).fillna(df_clean['beds'].median())

# Corroborar la sustituci√≥n
df_clean.isnull().sum()
```

```python
# Sustituci√≥n para los 5004 valores de Price.
## Formato de price, asegurando trabajar con string para luego reemplazar $ y ,
df_clean['price'] = df_clean['price'].astype(str).str.replace('[$,]', '', regex = True).astype(float)

# Media por roomtype
df_clean['room_type'].value_counts()
mean_prices = df_clean.groupby('room_type')['price'].mean()
print(mean_prices)

# Sustituci√≥n con la media por tipo de habitaci√≥n
df_clean['price'] = df_clean.apply(lambda row:mean_prices[row['room_type']] if pd.isna(row['price']) else row['price'], axis = 1)

# Corroborar
df_clean.isnull().sum()
```

```python
# Sustituci√≥n para estimated_revenue_l365d y sus 5004 na's.
''' 
Se identific√≥ que si estimated_occupancy_l365d es 0, por consecuencia estimated_revenue_l365d ser√° 0, as√≠ que se empezar√° por esa sustituci√≥n, adem√°s si hay estimated_occupancy, ser√° multiplicado por el precio
'''
df_clean.loc[df_clean['estimated_occupancy_l365d'] == 0, 'estimated_revenue_l365d'] = 0

df_clean['estimated_revenue_l365d'] = df_clean['estimated_revenue_l365d'].fillna(df_clean['price'] * df_clean['estimated_occupancy_l365d'])

df_clean.isnull().sum()
```

```python
# Sustituci√≥n para los Na's de review_scores_rating, review_scores_cleanliness, review_scores_location, review_scores_value and reviews_per_month
''' 
Se identific√≥ que si no se tiene estimated_occupancy_l365d, en algunos casos no se tiene review scores rating/cleanliness/location/value/month, entonces se empezar√° con esa primera sustituci√≥n.
'''
reviews_col = ['review_scores_rating', 'review_scores_cleanliness', 'review_scores_location', 'review_scores_value', 'reviews_per_month']
df_clean.loc[df_clean['number_of_reviews'] == 0, reviews_col] = 0

# Los na's restantes se llenaran utilizando el m√©todo forward fill. Solo quedaron 3 na's en review_scores_location y 5 na's en review_scores_value
df_clean['review_scores_value'] = df_clean['review_scores_value'].fillna(method = 'ffill')
df_clean['review_scores_location'] = df_clean['review_scores_location'].fillna(method = 'ffill')

# √öltima corraboraci√≥n
df_clean.isnull().sum()
```

```python
# Guardar df_clean a csv
df_clean.to_csv('Berlin_Clean.csv', index = False)
```
